# 03. If-Statements with User-Defined Functions

```python
# Create subset of df, first one million rows
df = ords_prods_merge[:1000000]
```

```python
# Define function that designates rows as a low-, mid-, or high-range product based on its price
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Check for most expensive item in df
df['prices'].max()
```

# 04. If-Statements with the loc() Function

```python
# loc() function will give same results but run faster
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Repeat process on entire df
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

# 05. If-Statements with For-Loops

```python
# Find on which day most orders take place
ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
# Loop through orders_day_of_week column
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# List produced from above loop
result
```

```python
# Combine result list with df as new column
ords_prods_merge['busiest_day'] = result
```

```python
ords_prods_merge['busiest_day'].value_counts(dropna=False)
```

```python
# Find what time of day most orders take place
ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

```python
# Loop to find busiest and least busy hours
# Top 6 hours are most orders, middle 12 are average orders, and bottom 6 are fewest orders
result3 = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12:
    result3.append("Most orders")
  elif value == 3 or value == 4 or value == 2 or value == 5 or value == 1 or value == 0:
    result3.append("Fewest orders")
  else:
    result3.append("Average orders")
```

```python
# Combine result list with df as new column
ords_prods_merge['busiest_period_of_day'] = result3
```
