# 05. Question 6

```python
# Create new column with the average amount each user spends on products
ords_prods_merge['avg_item_spend'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.average)
```

```python
ords_prods_merge.head()
```

```python
# Create spending_flag column that assignes spending label to user based on avg_item_spend
ords_prods_merge.loc[ords_prods_merge['avg_item_spend'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_item_spend'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# 06. Question 7

```python
# Create new column with the median days since prior order for each user
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
# Create order_freq_flag column that assignes how regular a customer is based on median_days_since_prior_order
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10)
                     & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'order_freq_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['order_freq_flag'].value_counts(dropna = False)
```
